## Nombre de Vente par Famille en fonction du Mois

```python
df_sell_by_familly = pd.DataFrame(df_raw)
df_sell_by_familly.drop(df_sell_by_familly.columns.difference(['FAMILLE','MOIS_VENTE', 'PRIX_NET']), 1, inplace=True)
df_sell_by_familly = df_sell_by_familly.groupby(["FAMILLE", "MOIS_VENTE"], as_index=True)["MOIS_VENTE"].count()

plot_df = df_sell_by_familly.unstack().T
plot_df.plot(grid=True, lw=2, marker='.', markersize=10, title='Nombre de Vente par Famille en fonction du Mois')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.show()
```

```python
df_mean_by_familly = pd.DataFrame(df_raw)
df_mean_by_familly.drop(df_mean_by_familly.columns.difference(['FAMILLE','MOIS_VENTE', 'PRIX_NET']), 1, inplace=True)
df_mean_by_familly = df_mean_by_familly.groupby(["FAMILLE", "MOIS_VENTE"], as_index=True)["PRIX_NET"].mean()

plot_df = df_mean_by_familly.unstack().T
plot_df.plot(grid=True, lw=2, marker='.', markersize=10, title='Evolution du prix par Famille en fonction du Mois')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

## Calc CA

```python
df_ca_by_familly = pd.DataFrame(df_raw)
df_ca_by_familly.drop(df_ca_by_familly.columns.difference(['FAMILLE','MOIS_VENTE', 'PRIX_NET']), 1, inplace=True)
df_ca_by_familly = df_ca_by_familly.groupby(["FAMILLE", "MOIS_VENTE"], as_index=True)["PRIX_NET"].sum()

plot_df = df_ca_by_familly.unstack().T
plot_df
plot_df.plot(grid=True, lw=2, marker='.', markersize=10, title='CA par Famille en fonction du Mois')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

plt.show()
```
